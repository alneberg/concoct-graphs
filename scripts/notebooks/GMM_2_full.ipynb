```python
import numpy as np
from sklearn.mixture import GMM
import itertools

import pylab as pl
import matplotlib as mpl
```

```python
from ggplot import ggplot, aes, geom_point
import pandas as p
```

```python
n, m = 300, 2

# generate random sample, two components
np.random.seed(0)
C = np.array([[0., -0.7], [3.5, .7]])
X = np.r_[np.dot(np.random.randn(n, 2), C),
          np.random.randn(n, 2) + np.array([2, 4])]

clf_full = GMM(n_components=2, covariance_type='full')
clf_full.fit(X)

clf_diag = GMM(n_components=2, covariance_type='diag')
clf_diag.fit(X)

Y_full = clf_full.predict(X)
Y_diag = clf_diag.predict(X)
```

```python
figure(num=None, figsize=(12, 12), dpi=120, facecolor='w', edgecolor='k')
splot = pl.subplot(111, aspect='equal')
color_iter = itertools.cycle (['r', 'b', 'g', 'c'])

for i, (mean, covar, color) in enumerate(zip(clf_full.means_, clf_full.covars_, color_iter)):
    v, w = np.linalg.eigh(covar)
    u = w[0] / np.linalg.norm(w[0])
    pl.scatter(X[Y_full==i, 0], X[Y_full==i, 1], 20, color=color, alpha = 0.25)
    angle = np.arctan(u[1]/u[0])
    angle = 180 * angle / np.pi # convert to degrees
    ell = mpl.patches.Ellipse (mean, v[0], v[1], 180 + angle, color=color)
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(0.5)
    splot.add_artist(ell)
pl.xlim(-13,10)

pl.show()
```

```python
figure(num=None, figsize=(12, 12), dpi=120, facecolor='w', edgecolor='k')
splot = pl.subplot(111, aspect='equal')
color_iter = itertools.cycle (['r', 'b', 'g', 'c'])

for i, (mean, covar_arr, color) in enumerate(zip(clf_diag.means_, clf_diag.covars_, color_iter)):
    pl.scatter(X[Y_diag==i, 0], X[Y_diag==i, 1], 20, color=color, alpha = 0.25)
    ell = mpl.patches.Ellipse (mean, covar_arr[0], covar_arr[1], color=color)
    ell.set_clip_box(splot.bbox)
    ell.set_alpha(0.5)
    splot.add_artist(ell)
pl.xlim(-13,10)

pl.show()
```
